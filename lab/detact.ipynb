```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# อ่านภาพสีด้วย OpenCV
img1_bgr = cv2.imread('img/box1.png', cv2.IMREAD_COLOR)

img1_rgb = cv2.cvtColor(img1_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img1_rgb)
plt.show()
```

```python
# กำหนดขอบ ให้พื้นที่นอกเป็น 0 
img1_bgr[350:,:] = (0,0,0)

plt.imshow(img1_bgr)
```

```python
img1_blur = cv2.GaussianBlur(img1_bgr, (7,7), 0)
img1_hsv = cv2.cvtColor(img1_blur, cv2.COLOR_BGR2HSV)

plt.imshow(img1_hsv)
plt.show()
```

```python
# Define range for red color in HSV (two ranges)
lower_red_1 = np.array([0, 50, 50], dtype=np.uint8)
upper_red_1 = np.array([10, 255, 255], dtype=np.uint8)
lower_red_2 = np.array([170, 50, 50], dtype=np.uint8)
upper_red_2 = np.array([180, 255, 255], dtype=np.uint8)

mask1 = cv2.inRange(img1_hsv, lower_red_1, upper_red_1)
mask2 = cv2.inRange(img1_hsv, lower_red_2, upper_red_2)
mask = cv2.bitwise_or(mask1, mask2)

plt.imshow(mask, cmap='gray')
plt.show()
```

```python
mask = mask[150:350,300:400]

plt.imshow(mask,cmap="gray")
plt.show()
```
